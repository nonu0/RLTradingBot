```python
import seaborn as sns
import pandas as pd
data = pd.read_csv(r'C:\Users\Administrator\work\RLTrading\Bot\data\processed\processed.csv',index_col=['time'])
df = pd.DataFrame(data)
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

Closing Prices vs Time

```python
import matplotlib.pyplot as plt
sns.histplot(df,x='time',y='close')
plt.figure(figsize=(700,600))
plt.show()
```

Tick Volume vs Time

```python
import matplotlib.pyplot as plt
sns.boxplot(df,x='time',y='tick_volume')
plt.figure(figsize=(700,600))
plt.show()
```

```python
sns.barplot(df,x='time',y='tick_volume')
```

Log Returns vs Time

```python
sns.scatterplot(df,x='time',y='log_returns')
sns.rugplot(df,x='time',y='log_returns')
```
